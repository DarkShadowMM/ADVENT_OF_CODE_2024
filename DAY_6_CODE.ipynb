```python
with open("day_6_input.txt") as file:
    lines=file.read().strip().split("\n")
    print(lines)
```

```python
n=len(lines)#setir
m=len(lines[0])#sutun
for i in range(n):
    for j in range(m):
        if lines[i][j]=="^":
            print(lines[i][j])
            row = i
            col=j
            print(i,j)
            break
```

```python
path=[]
while True:
    path.append([row,col])
    break
```

```python
flag_guard=False 

for i in range(n): 
    for j in range(m): 
        if lines[i][j]=="^": 
            flag_guard=True
            break
    if flag_guard: 
        break
        

path=set() 
dire=0
dd=[[-1,0], [0,1], [1,0], [0,-1]]

while True: 
    path.add((i,j))
    
    new_i=i+dd[dire][0]
    new_j=j+dd[dire][1]
    
    if not (0<=new_i<n and 0<=new_j<n): 
        break
        
    if lines[new_i][new_j]=="#": 
        dire=(dire+1)%4
    
    else: 
        i,j=new_i, new_j
```
