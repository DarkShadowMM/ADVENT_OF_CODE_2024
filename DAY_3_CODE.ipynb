```python
import re

with open("day 3.txt","r") as file:
    s=file.read()    
matches=re.findall(r"mul\((\d+),(\d+)\)",s)
    
result=[]
for x, y in matches:
    print(f"mul({x},{y})  ->  x={x}, y={y}")
    x,y=int(x),int(y)
    res=x*y
    result.append(res)
print( sum(result))

    
```

```python
import re

with open("day 3.txt", "r") as file:
    s = file.read()
matches=re.findall(r"do\(\)|don't\(\)|mul\(\d+,\d+\)", s)
enabled=True
result=[]
for match in matches:
    if match=="do()":
        enabled=True
    elif match=="don't()":
        enabled=False
    elif "mul(" in match and enabled:
        x,y=map(int, re.findall(r"\d+",match))
        print(f"mul({x},{y}) -> x={x},y={y}")
        result.append(x*y)

print(sum(result))
```
